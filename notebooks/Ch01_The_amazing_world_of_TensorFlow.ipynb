## Konsep teoritis utama

### 1) Machine Learning vs Deep Learning
- **Machine Learning (ML)**: melatih model agar memetakan input → output berdasarkan contoh data.
- **Deep Learning (DL)**: subset ML yang menggunakan jaringan saraf berlapis untuk belajar representasi fitur bertingkat.

Yang relevan untuk chapter berikut:
- DL biasanya membutuhkan banyak komputasi, sehingga akselerasi (GPU/TPU) sering krusial.
- Banyak praktik penting (normalisasi data, pemilihan loss/optimizer, evaluasi) baru terlihat jelas setelah kita mulai menulis dan menjalankan eksperimen.

### 2) Kenapa TensorFlow populer untuk DL
TensorFlow menargetkan dua kebutuhan sekaligus:
- **Produktivitas pengembangan**: API tingkat tinggi (Keras) untuk membangun model cepat.
- **Performa**: komputasi numerik yang dapat berjalan di CPU/GPU/TPU, serta konsep graph execution untuk optimasi tertentu.

### 3) Kekuatannya vs keterbatasannya (secara praktis)
- Kuat ketika kamu membangun dan melatih model DL (terutama untuk visi, bahasa, dan data skala besar).
- Bisa terasa berlebihan atau kurang nyaman untuk analisis data tabular kecil yang eksploratif, atau pipeline teks yang sangat rumit bila dibandingkan dengan tool yang lebih spesifik.
